```python
import pandas as pd
import matplotlib.pyplot as plt
import textwrap
import os

OUT_DIR = "eda_images"
os.makedirs(OUT_DIR, exist_ok=True)
df = pd.read_csv("/content/drive/MyDrive/StackOverFlow/survey_cleaned_base.csv")

plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 10,
    'axes.titlesize': 12,
    'axes.labelsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'legend.fontsize': 9,
    'figure.titlesize': 14
})

def wrap_index(series, width=35):
    series.index = [textwrap.fill(str(idx), width) for idx in series.index]
    return series

def explode_multi(series, sep=";"):
    return series.dropna().str.split(sep).explode().str.strip().value_counts()

targets = [
    ("LanguageHaveWorkedWith", "Топ-15 языков программирования (используемых в работе)"),
    ("DatabaseHaveWorkedWith", "Топ-15 баз данных (используемых в работе)"),
    ("AISearchDevHaveWorkedWith", "Топ-15 AI-инструментов, используемых в разработке")
]

for col, title in targets:
    if col in df.columns:
        counts = explode_multi(df[col]).head(15)
        fig, ax = plt.subplots(figsize=(10, 6))
        wrap_index(counts).plot.barh(ax=ax, color="#5B9BD5")
        ax.invert_yaxis()
        ax.set_xlabel("Упоминаний")
        ax.set_title(title)
        fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, f"05_{col}_top15.png"), dpi=300)
        plt.close(fig)
```
